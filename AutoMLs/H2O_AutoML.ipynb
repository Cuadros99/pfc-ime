```python
train[y] = train[y].asfactor()
test[y] = test[y].asfactor()

aml = H2OAutoML(max_runtime_secs=1800, seed=42, nfolds=5, stopping_metric="AUCPR")
aml.train(x=x, y=y, training_frame=train)

lb = aml.leaderboard
lb_df = lb.as_data_frame()

best_model = aml.leader

perf = best_model.model_performance(test)
```

```python
results = []

# Avaliar cada modelo na leaderboard
for model_id in lb['model_id'].as_data_frame().iloc[:, 0]:
    model = h2o.get_model(model_id)
    perf = model.model_performance(test)

    # Verificar a AUC PR
    pr_auc = perf.aucpr()

    # Obter precisão e recall no limiar padrão
    precision = perf.precision()[0][1] if len(perf.precision()) > 0 else None
    recall = perf.recall()[0][1] if len(perf.recall()) > 0 else None

    # Obter o F2 Score, garantindo que ele existe
    f2 = perf.F2()
    f2_value = f2[0][1] if len(f2) > 0 else None

    # Obter a matriz de confusão e calcular FPR
    cm = perf.confusion_matrix().to_list()
    tp = cm[0][0] if len(cm) > 0 and len(cm[0]) > 0 else None
    fn = cm[0][1] if len(cm) > 0 and len(cm[0]) > 1 else None
    fp = cm[1][0] if len(cm) > 1 and len(cm[1]) > 0 else None
    tn = cm[1][1] if len(cm) > 1 and len(cm[1]) > 1 else None

    fpr = fp / (fp + tn) if (fp is not None and tn is not None and (fp + tn) != 0) else None

    # Adicionar resultados à lista
    results.append({
        "Modelo": model_id,
        "PR-AUC": pr_auc,
        "Precisão": precision,
        "Recall": recall,
        "F2 Score": f2_value,
        "FPR": fpr
    })

# Converter resultados para DataFrame
results_df = pd.DataFrame(results)

```

```python
from sklearn.metrics import recall_score
def calculate_recall(model, test_data, y_true):
    predictions = model.predict(test_data).as_data_frame()
    y_pred = predictions['predict'].values
    recall = recall_score(y_true, y_pred)
    return recall

# Inicializar uma lista para armazenar os resultados
results = []

# Calcular o recall para cada modelo no leaderboard
for model_id in lb_df['model_id']:
    model = h2o.get_model(model_id)
    y_true = test[y].as_data_frame().values.flatten()
    recall = calculate_recall(model, test, y_true)
    results.append({'model_id': model_id, 'recall': recall})

# Converter os resultados para um DataFrame do Pandas
recall_df = pd.DataFrame(results)
```

```python
recall_df = recall_df.sort_values(by='recall', ascending=False)
recall_df
```

```python
from sklearn.metrics import precision_score
def calculate_precision(model, test_data, y_true):
    predictions = model.predict(test_data).as_data_frame()
    y_pred = predictions['predict'].values
    precision = precision_score(y_true, y_pred)
    return precision

# Inicializar uma lista para armazenar os resultados
results = []

# Calcular a precisão para cada modelo no leaderboard
for model_id in lb_df['model_id']:
    model = h2o.get_model(model_id)
    y_true = test[y].as_data_frame().values.flatten()
    precision = calculate_precision(model, test, y_true)
    results.append({'model_id': model_id, 'precision': precision})

# Converter os resultados para um DataFrame do Pandas
results_df = pd.DataFrame(results)

# Ordenar o DataFrame pela métrica de precisão em ordem decrescente
precision_df = results_df.sort_values(by='precision', ascending=False)

precision_df
```

```python
from sklearn.metrics import fbeta_score

def calculate_f2(model, test_data, y_true):
    predictions = model.predict(test_data).as_data_frame()
    y_pred = predictions['predict'].values
    f2 = fbeta_score(y_true, y_pred, beta=5)
    return f2

# Inicializar uma lista para armazenar os resultados
results = []

# Calcular o F5 para cada modelo no leaderboard
for model_id in lb_df['model_id']:
    model = h2o.get_model(model_id)
    y_true = test[y].as_data_frame().values.flatten()
    f2 = calculate_f2(model, test, y_true)
    results.append({'model_id': model_id, 'F2': f2})

# Converter os resultados para um DataFrame do Pandas
results_df = pd.DataFrame(results)

# Ordenar o DataFrame pelo F5 em ordem decrescente
f2_df = results_df.sort_values(by='F2', ascending=False)
```

```python
f2_df
```

```python
from sklearn.metrics import confusion_matrix

def calculate_fpr(model, test_data):
    predictions = model.predict(test_data).as_data_frame()
    y_true = test_data[y].as_data_frame().values.flatten()
    tn, fp, fn, tp = confusion_matrix(y_true, predictions['predict']).ravel()
    fpr = fp / (fp + tn)
    return fpr

# Inicializar uma lista para armazenar os resultados
results = []

# Calcular o FPR para cada modelo no leaderboard
for model_id in lb_df['model_id']:
    model = h2o.get_model(model_id)
    fpr = calculate_fpr(model, test)
    results.append({'model_id': model_id, 'fpr': fpr})

# Converter os resultados para um DataFrame do Pandas
results_df = pd.DataFrame(results)

# Ordenar o DataFrame pelo FPR em ordem crescente
fpr_df = results_df.sort_values(by='fpr', ascending=True)
```
